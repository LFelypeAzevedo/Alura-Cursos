```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df.set_index('País',inplace=True)
```

```python
anos=list(map(str, range(1980,2014)))
```

```python
df_comp = df.loc[['Brasil', 'Argentina'], anos].T
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(df_comp.index,df_comp['Brasil'],label='Brasil')
plt.plot(df_comp.index,df_comp['Argentina'],label='Argentina')
plt.title('Comparação de Imigrações para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.legend()
plt.xticks(list(map(str,range(1980,2014,5))))
plt.show()
```
