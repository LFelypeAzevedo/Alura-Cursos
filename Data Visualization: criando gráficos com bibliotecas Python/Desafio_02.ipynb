```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_vendas = pd.DataFrame(vendas_2022,index=lojas)
```

```python
df_vendas.loc['A']
```

```python
df_vendas.values
```

```python
fig,axs = plt.subplots(2,2, figsize=(16,9))
fig.subplots_adjust(hspace=0.5,wspace=0.2)
fig.suptitle('Vendas por mês das lojas em 2022')

i = 0
for ax in axs.flat:
  ax.plot(df_vendas.loc[lojas[i]])
  ax.set_title('Vendas da Loja {0}'.format(lojas[i]))
  ax.set_xlabel('Mês')
  ax.set_ylabel('Vendas')
  ax.grid()
  i+=1

ymin = 0
ymax = (np.ceil(np.amax(df_vendas.values)/100)+1)*100

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

plt.show()
```
